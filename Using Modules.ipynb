```python
import numpy as np
import matplotlib.pyplot as matplot
```

```python
grades = np.random.normal(80,20,1000) #genelde 80'e yakın, standart sapması 10 olan 1000 tane random değişken oluşturduk.
```

```python
np.mean(grades)
```

Bütün random değişkenlerin ortalamasını bulduk.

```python
matplot.hist(grades,100) # notların 100 tane barla histogram grafiğine çevrilmiş hali
matplot.show()
```
